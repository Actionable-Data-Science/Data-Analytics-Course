### Curiosity Exercise
#### Which author uses a more sophisticated language? 

Think:

   - The use of long words? 

Do:
1. Load Melville’s Moby Dick, and Shakespeare’s Hamlet and Macbeth
2. Compare the size of vocabularies
3. Check the use of long words


```python
# Let's simply compare the vocabularies of the books of different authors

moby_book = nltk.Text(nltk.corpus.gutenberg.words('melville-moby_dick.txt'))
print("No of words in Moby Dick: ",len(moby_book))
print("No of unique words in Moby Dick: ",len(set(moby_book)))

macbeth_book = nltk.Text(nltk.corpus.gutenberg.words('shakespeare-macbeth.txt')) # "Macbeth" (1605-1606)
print("No of words in Macbeth: ", (len(macbeth_book)))
print("No of unique words in Macbeth: ",len(set(macbeth_book)))

hamlet_book = nltk.Text(nltk.corpus.gutenberg.words('shakespeare-hamlet.txt')) # Hamlet" (1600-1601)
print("No of words in Hamlet: ", len(hamlet_book))
print("No of unique words in Hamlet: ", len(set(hamlet_book)))

#print(len(set(hamlet).intersection(set(macbeth))))
```

```python
## Type-Token Ratio (Normalize the number of unique words on the lenth of text)

print("Type-Token Ratio of Moby Dick: ", len(set(moby_book)) / len(moby_book))
print("Type-Token Ratio of Macbeth: ", len(set(macbeth_book)) / len(macbeth_book))
```

```python
# Let's check the number of words longer than X characters
V = set(moby_book)
moby_long_words = [w for w in V if len(w) > 10]
#print(sorted(long_words))
print("No of unique long-words in Moby Dick: ", len(moby_long_words))

V = set(macbeth_book)
macbeth_long_words = [w for w in V if len(w) > 10]
#print(sorted(long_words))
print("No of unique long-words in Macbeth: ", len(macbeth_long_words))

V = set(hamlet_book)
hamlet_long_words = [w for w in V if len(w) > 10]
#print(sorted(long_words))
print("No of unique long-words in Hamlet: ", len(hamlet_long_words))

print("\nFirst 20 long words in Hamlet: ", hamlet_long_words[:20])
print("\nFirst 20 long words in Moby Dick: ",moby_long_words[0:20])
```

```python
# some set operations to see the different words used
print(len(set(hamlet_long_words) - set(moby_long_words)))
print(len(set(macbeth_book) - set(hamlet_book)))
```

### Count, Frequency, and Dispersion

```python
# Number of ocurrence of a specific word
# 
print(alice_book.count("Alice"))

## Count the number of the mentions of Queen.
print(alice_book.count("Queen"))


## is it case-sensitive? How to check? 
print(alice_book.count("alice"))

```

```python
# Check for the count of words "Moby" and "Ahab" in Moby Dick
# 
print(moby_book.count("Moby"))
print(moby_book.count("Ahab"))
print(moby_book.count("whale"))
```

```python
# FREQUENCY DISTRIBUTION OF WORDS
#
from nltk.probability import FreqDist
fdist = FreqDist(alice_book)
print(fdist)
fdist.most_common(20)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))

# We may plot the top 70 most common words in “Alice in Wonderland” by creating a plot:
fdist = nltk.FreqDist(alice_book)
fdist.plot(70, cumulative=False)
```

```python
plt.figure(figsize=(12,6))

# We may plot the top 70 most common words in “Alice in Wonderland” by creating a plot:
fdist = nltk.FreqDist(alice_book)
fdist.plot(70, cumulative=True)
```

```python
# Check for the frequency distribution of words in Moby Dick
# 
from nltk.probability import FreqDist
fdist_mob = FreqDist(moby_book)
print(fdist_mob)
print(fdist_mob.most_common(20))

# Plot the frequency distribution in Moby Dick
# 
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))


fdist_mob.plot(70, cumulative=False)
```

The most frequent words are not suprisingly stop words such as _"the"_ and _"and"_, and punctuations. This shows us that we should somehow deal with them to get meaningful results in our analysis.

```python
# Dispersion Plot
# However, we can also determine the location of a word in the text: how many words from the beginning it appears. 
# This positional information can be displayed using a dispersion plot.

alice_book.dispersion_plot(["Alice", "Rabbit", "Hatter", "Queen"])
```

Notice on the dispersion plot that Alice is mentioned all along the text while Queen only appears at a later stage

```python
# Plot the dispersion of words "monstrous", "whale", "Moby", and "Ahab" in Moby Dick
# 
moby_book.dispersion_plot(["captain", "Ahab", "whale", "monstrous"])
```

```python
## To implement your own dispersion plot. 
import nltk
from nltk.corpus import webtext
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
 
words = [ 'Alice', 'Bob', 'billion']

points = [(x, y) for x in range(len(textTC))
          for y in range(len(words)) if textTC[x] == words[y]]
 
if points:
    x, y = zip(*points)
else:
    x = y = ()

plt.plot(x, y, "rx", scalex=.1)
plt.yticks(range(len(words)), words, color="b")
plt.ylim(-1, len(words))
plt.xlim(0,len(textTC))
plt.title("Lexical Dispersion Plot")
plt.xlabel("Word Offset")
plt.show()
```

### Context, Concordance, and Similarity

```python
# https://www.nltk.org/book/ch01.html
#  A concordance view shows us every occurrence of a given word, together with some context. 
# Here we look up the word monstrous in Moby Dick.

moby_book.concordance("monstrous")

## To see the details:
# moby_book.concordance?
```

```python
# Look up the word "monstrous" in Austen's Sense and Sensibility 
austen_book = nltk.Text(nltk.corpus.gutenberg.words('austen-sense.txt')) 
austen_book.concordance("monstrous")
```

```python
# A concordance permits us to see words in context. 
# What other words appear in a similar range of contexts? We can find out by function "similar"
# Documentation: http://www.nltk.org/_modules/nltk/text.html
# Distributional similarity: find other words which appear in the same contexts as the specified word; list most similar words first.

print("Melville's use of the word monstrous: ")
moby_book.similar("monstrous")
print("\n")
print("Austen's use of the word monstrous: ")
austen_book.similar("monstrous")
```
